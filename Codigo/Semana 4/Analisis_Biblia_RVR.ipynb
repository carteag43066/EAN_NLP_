```python
def extract_words_from_pdf(pdf_path):
    # Cargar el archivo PDF
    reader = PdfReader(pdf_path)
    text = ""

    # Iterar sobre las páginas y extraer el texto
    for page in reader.pages:
        text += page.extract_text()

    # Dividir el texto en palabras
    words = text.split()

    return words
```

```python
import pandas as pd

def extraer_todos_los_libros_en_dataframe(pdf_path, libros_paginas):
    libros = list(libros_paginas.keys())
    paginas_inicio = list(libros_paginas.values()) + [None]  # Límite final para el último libro
    data = []

    with open(pdf_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        total_pages = len(reader.pages)
        
        for i, libro in enumerate(libros):
            inicio = paginas_inicio[i]
            fin = paginas_inicio[i + 1] - 1 if paginas_inicio[i + 1] else total_pages
            texto_libro = ""
            
            for page_num in range(inicio - 1, fin):
                page = reader.pages[page_num]
                texto_libro += page.extract_text()
            
            # Añadir libro y texto al dataset
            data.append({'Libro': libro, 'Texto': texto_libro})
    
    # Crear DataFrame
    df = pd.DataFrame(data)
    return df
```

```python
# Crear DataFrame con todos los libros
df_libros = extraer_todos_los_libros_en_dataframe(pdf_path, libros_paginas)

# Mostrar los primeros registros del DataFrame
print(df_libros.head())
```

```python
import re
import html
import nltk
from nltk.corpus import stopwords
from unidecode import unidecode


# Descargar stopwords en español
nltk.download('stopwords')

def limpiar_data(data):
    # Decodificar entidades HTML
    data = html.unescape(data)

    # Eliminar etiquetas HTML
    data = re.sub(r'<.*?>', '', data)

    # Convertir a minúsculas
    data = data.lower()

    # Reducir letras repetidas
    data = re.sub(r'(.)\1{2,}', r'\1\1', data)

    # Eliminar palabras repetidas
    data = re.sub(r'\b(\w+)( \1\b)+', r'\1', data)

    # Eliminar menciones
    data = re.sub(r'@\w+', '', data)

    # Eliminar URLs
    data = re.sub(r'http\S+|www\.\S+', '', data)

    # Eliminar puntuación pero conservar caracteres especiales en español
    data = re.sub(r'[^a-zA-ZáéíóúñüÁÉÍÓÚÑÜ]', ' ', data)

    # Eliminar espacios en blanco adicionales
    data = re.sub(r'\s+', ' ', data).strip()

    # Eliminar stopwords en español
    stopwords_es = set(stopwords.words('spanish'))
    data = ' '.join([word for word in data.split() if word not in stopwords_es])

    # Eliminar tildes y diacríticos
    data = unidecode(data)

    return data
```

```python
# Aplicar la función a cada elemento de la lista
data_limpio = [limpiar_data(texto) for texto in data]

# Mostrar el resultado
print(data_limpio[:20])
```
